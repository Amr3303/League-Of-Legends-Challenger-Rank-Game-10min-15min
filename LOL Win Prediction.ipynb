```python
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
from tensorflow.python.keras.layers import Dense

pd.set_option('display.max_columns', None)
```

##### Loading the dataset

```python
df = pd.read_csv("./high_diamond_ranked_10min.csv")
df
```

```python
df.shape
```

```python
df.drop('gameId', axis= 1, inplace= True)
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.sample(frac= 1).reset_index(drop= True)
df
```

```python
X = df.drop('blueWins', axis= 1)
X
```

```python
y = df['blueWins']
y
```

```python
# see the correlation

import plotly.express as px
px.imshow(X.corr())
```

```python
train_test_split = 0.85

num_examples = X.shape[0]

num_train_examples = int(np.floor(num_examples * train_test_split))
num_test_examples = int(np.ceil(num_examples - num_train_examples))

print(num_examples)
print(num_train_examples)
print(num_test_examples)
```

```python
optimizer = tf.keras.optimizers.Adam(
    learning_rate= 0.001, beta_1= 0.9, beta_2= 0.999, epsilon= 1e-07, amsgrad= False,
    name= "Adam"
)
```

```python
inputs = tf.keras.Input(shape= (38, ))
x = tf.keras.layers.Dense(16, activation= tf.nn.relu)(inputs)
x = tf.keras.layers.Dense(16, activation= tf.nn.relu)(x)
outputs = tf.keras.layers.Dense(2, activation= tf.nn.softmax)(x)

model = tf.keras.Model(inputs = inputs, outputs = outputs)
model.summary()
```

```python
model.compile(
    optimizer= optimizer, loss = tf.keras.losses.SparseCategoricalCrossentropy(), metrics = ['accuracy']
)
```

```python
y.shape
```

```python
y.iloc[:num_train_examples]
```

```python
X_train = X.iloc[0:num_train_examples, :]
Y_train = y.iloc[0:num_train_examples]

X_test = X.iloc[num_train_examples: , :]
Y_test = y.iloc[num_train_examples:]
```

```python
BATCH_SIZE = 64
EPOCHS = 300
```

```python
model.fit(
    x = X_train,
    y = Y_train,
    batch_size= BATCH_SIZE,
    epochs= EPOCHS,
    verbose = 1,
    validation_split = 0.2,
    shuffle= True
)
```

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(f"test loss {loss}\ntest Accuracy: {accuracy}")
```
